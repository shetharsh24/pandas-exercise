## Importing necessary libraries

```python
import numpy as np
import pandas as pd
import glob
```

#### Merging 12 files of data into a single file

```python
df = pd.concat(map(pd.read_csv,glob.glob('Sales_Data/*.csv')))
```

```python
df.to_csv('Final_Sales_1.csv',index=False)
```

## Additional column for operations

### Add the month column

```python
df.head(5)
```

```python
nan_df = df[df.isna().any(axis=1)]
nan_df.head()
df = df.dropna(how='all')
df.head(5)
```

```python
df['Month'] = df['Order Date'].str[0:2]
df.head(5)
```

```python
df = df[df['Order Date'].str[0:2] != 'Or']
```

```python
df['Month'] = df['Month'].astype('int32')
df.head()
```

```python
df.head(40)
```

```python
df['Quantity Ordered'] = df['Quantity Ordered'].astype('int32')
```

```python
df.head(40)
```

```python
df['Price Each'] = pd.to_numeric(df['Price Each'])
```

```python
df['Sales'] = df['Quantity Ordered']*df['Price Each']
df.head(5)
```

## Question 1 - What was the best month for sales and how much was earned?

```python
result = df.groupby('Month').sum()
result
```

```python
import matplotlib.pyplot as plt
```

```python
months = range(1,13)

plt.bar(months,result['Sales'])
plt.yscale('log')
plt.xticks(months)
plt.xlabel('Months')
plt.ylabel('Sales in USD')
plt.show()
```

## Question 2: What city sold the most product?

```python
def get_address(address):
    return address.split(',')[2].split(' ')[1]


df['City'] = df['Purchase Address'].apply(lambda x:x.split(',')[1] + ' ' + get_address(x))
```

```python
df.head(5)
```

```python
res = df.groupby('City').sum()
res
```

```python
cities = [city for city,af in df.groupby('City')]

plt.bar(cities,res['Sales'])
plt.xticks(cities,rotation='vertical')
plt.xlabel('cities')
plt.ylabel('Sales in USD')
plt.show()
```

## Question 4: What products are most often sold together?

```python
fd = df[df['Order ID'].duplicated(keep=False)]
fd['Grouped'] = fd.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
fd = fd[['Order ID','Grouped']].drop_duplicates()
fd.head()
```

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()

for row in fd['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
for key,value in count.most_common(10):
    print(key,value)
```

## Question 5: What product sold the most? Why do you think it did?

```python
product_groupby = df.groupby('Product')
quantity_ordered = product_groupby.sum()['Quantity Ordered']
```

```python
products = [product for product,af in product_groupby]
plt.bar(products,quantity_ordered)
plt.xticks(products,rotation='vertical')
plt.ylabel('Quantity Ordered')
plt.xlabel('Products')
plt.show()
```

```python
prices = df.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered,color='g')
ax2.plot(products,prices,color='b')

ax1.set_xlabel('Products')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Prices', color='b')
ax1.set_xticklabels(products,rotation='vertical')

plt.show()
```
